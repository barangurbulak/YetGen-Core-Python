random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import time
import random
class Kumanda:
    def __init__(self,marka="Vestel"):
        self.marka=marka
        self.ses=10
        self.kanallar=["Now","Star","Show","HalkTV","KanalD"]
        self.acik=False
    def tv_ac(self):
        if not self.acik:
            self.acik=True
            print("TV açılıyor..")
            print("TV açıldı.")
        else:
            print("TV zaten açık")

    def tv_kapat(self):
        if self.acik:
            self.acik=False
            print("TV kapatılıyor..")
            print("TV kapatıldı.")
        else:
            print("TV zaten kapalı")
    def ses(self):
        a=int(input("sesin kaç olmasını istiyorsunuz?"))
        self.ses=a
        
        print(f"ses {self.ses} olarak ayarlanmıştır")
    def kanal_ekle(self):
        b=input("eklemek istediğiniz kanalı yazınız:")
        self.kanallar.append(b)
        
        print(f"{b} kanalı eklendi")
    def kanal_sayısı(self):
        print("Kanal sayısı kontrol ediliyor..")
       
        print(f"kanal sayısı: {len(self.kanallar)}")
    def random(self):
        print("Rastgele bir kanal seçiliyor..")
        1
        print(f"{random.choice(self.kanallar)} açıldı!")
    def info(self):
        print(f""" 
              Kanal Bilgileri:
              Marka: {self.marka}
              Ses: {self.ses}
              Kanal Listesi: {self.kanallar}
              """)
def main():
        kumanda = Kumanda()
    
        while True:
            print("\n" + "*"*20)
            print(" TELEVİZYON KUMANDASI ")
            print("*"*20)
            print("1. TV Aç")
            print("2. TV Kapat")
            print("3. Ses Ayarları")
            print("4. Kanal Ekle")
            print("5. Kanal Sayısını Öğren")
            print("6. Rastgele Kanala Geç")
            print("7. Televizyon Bilgileri")
            print("q. Çıkış")
            print("\n" + "*"*20)
        

            secim=input("Seçiminizi giriniz:").lower()
            if secim == '1':
                kumanda.tv_ac()
            elif secim == '2':
                kumanda.tv_kapat()
            elif secim == '3':
                kumanda.ses()
            elif secim == '4':
                kumanda.kanal_ekle()
            elif secim == '5':
                kumanda.kanal_sayısı()
            elif secim == '6':
                kumanda.random()
            elif secim == '7':
                kumanda.info()
            else:
                print("Geçersiz seçim! Lütfen 1-7 arası bir sayı veya q girin.")
main()
               
```
